```python
import pandas as pd
df = pd.read_csv("/Finance.csv")
df.head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.naive_bayes import BernoulliNB, MultinomialNB
```

```python
custom_stopwords = ['this', 'is', 'the']
```

```python
vectorizer1=CountVectorizer(binary=True, stop_words=custom_stopwords) #to see if the word is there or not
vectorizer2=CountVectorizer(binary=False,stop_words=custom_stopwords) #Number of times the word has occurred
```

```python
x1=vectorizer1.fit_transform(df.Text)
x2=vectorizer2.fit_transform(df.Text)
```

```python
y=df.Label
```

```python
from sklearn.model_selection import train_test_split
xtrain1,xtest1,ytrain,ytest=train_test_split(x1,y,test_size=0.25,random_state=42)
```

```python
xtrain2,xtest2,ytrain,ytest=train_test_split(x2,y,test_size=0.25,random_state=42)
```

```python
bnb=BernoulliNB()
mnb=MultinomialNB()
```

```python
 bnb.fit(xtrain1,ytrain)
```

```python
mnb.fit(xtrain2,ytrain)
```

```python
y_pred1 = bnb.predict(xtest1)
```

```python
ypred2 = mnb.predict(xtest2)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,y_pred1)
```

```python
accuracy_score(ytest,ypred2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
custom_stopwords = ['this', 'is', 'the']
vectorizer3=TfidfVectorizer(stop_words=custom_stopwords)
```

```python
x3=vectorizer3.fit_transform(df.Text)
from sklearn.model_selection import train_test_split
xtrain3,xtest3,ytrain,ytest=train_test_split(x3,y,test_size=0.25,random_state=42)
```

```python
mnb=MultinomialNB()
mnb.fit(xtrain3,ytrain)
```

```python
feature_names = vectorizer3.get_feature_names_out()
print(feature_names)
```

```python
pred1 = mnb.predict(xtest3)
```

```python
accuracy_score(ytest,pred1)
```

```python
# User input for prediction
user_input = input("Enter a sentence for sentiment prediction: ")

# Transform the user input
user_input_transformed2 = vectorizer2.transform([user_input])

# Predict sentiment
user_pred2 = mnb.predict(user_input_transformed2)

# Output the predictions
print(f"Prediction with CountVectorizer (binary=False): {user_pred2[0]}")
```
